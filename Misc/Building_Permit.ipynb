```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df = gpd.read_file("../../../Data/Original_dataset/BuildingPermits_2019_02/BuildingPermits.shp");df
```

```python
df = df[df["typeofbldg"].isin(['New', 'Addition', 'New Bldg', 'Addition to existing Bldg'])];df
```

```python
df["issuedate"] = df["issuedate"].str[:4]
```

```python
count_summary = df['issuedate'].value_counts().sort_index()
count_summary
```

```python
# Count occurrences of each unique year
count_summary = df['issuedate'].value_counts().sort_index()
count_summary = count_summary.loc["2000":"2023"]

# Plot the line chart
plt.figure(figsize=(25, 6))
plt.plot(count_summary.index, count_summary.values, marker='o', linestyle='-', color = "black")
plt.xlabel('Year')
plt.ylabel('Building Permit')
plt.grid(True)
plt.show()
```
