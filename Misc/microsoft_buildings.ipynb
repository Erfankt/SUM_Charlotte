```python
import geopandas as gpd
```

```python
# 1. Load the full NC file (this may take time)
buildings = gpd.read_file("../../../../Data/Original_dataset/Archive/Buildings_footprint_nc_microsoft/NorthCarolina.geojson/NorthCarolina.geojson")

# 2. Load a Mecklenburg County boundary (from TIGER/Line, or your own shapefile)
meck = gpd.read_file("https://www2.census.gov/geo/tiger/TIGER2023/COUNTY/tl_2023_us_county.zip")
meck = meck[meck["NAME"] == "Mecklenburg"]

# 3. Reproject to the same CRS
meck = meck.to_crs(buildings.crs)

# 4. Spatial clip (only footprints inside the county)
meck_buildings = gpd.overlay(buildings, meck, how="intersection")

# 5. Save result
meck_buildings.to_file("Mecklenburg_BuildingFootprints.geojson", driver="GeoJSON")

print(f"{len(meck_buildings)} buildings saved.")
```
