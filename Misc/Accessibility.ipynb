```python
import geopandas as gpd, pandas as pd
```

```python
ABT = gpd.read_file("../../../Data/Final_dataset/ABT/ABT.gpkg", layer="subdivisions")
groceries_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Groceries/Grocery_Stores/Grocery_Stores_(points).shp").to_crs(ABT.crs)
meck_bo = gpd.read_file("../../../Data/Original_dataset/Archive/mecklenburgcounty_boundary/MecklenburgCounty_Boundary.shp").to_crs(ABT.crs)
groceries_dataset = groceries_dataset[groceries_dataset.within(meck_bo.geometry.iloc[0])]
bus_stop_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Transit/Bus_Stops_With_Frequency_HLT/Bus_Stops_With_Frequency_HLT.shp").to_crs(ABT.crs)
lynx_blue_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Transit/LYNX_Blue_Line_Stations/LYNX_Blue_Line_Stations.shp").to_crs(ABT.crs)
lynx_gold_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Transit/LYNX_Gold_Line_Stops/LYNX_Gold_Line_Stops.shp").to_crs(ABT.crs)
lynx_silver_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Transit/LYNX_Silver_Line_Stations_Proposed/LYNX_Silver_Line_Stations_Proposed.shp").to_crs(ABT.crs)
lynx_redline_dataset = gpd.read_file("../../../Data/Original_dataset/Archive/Transit/LYNX_Red_Line_Stations_Proposed/LYNX_Red_Line_Stations_Proposed.shp").to_crs(ABT.crs)
```

```python
#merging transit stations
transit_stations = pd.concat([bus_stop_dataset, lynx_blue_dataset, lynx_gold_dataset, lynx_silver_dataset, lynx_redline_dataset], ignore_index=True)
```

```python
transit_stations_4326 = transit_stations.to_crs(epsg=4326);transit_stations_4326
```

```python
transit_stations_4326[["StopID","StopDesc","LONGITUDE","LATITUDE","geometry"]].to_file("../../../Data/Original_dataset/transit_stations_4326.shp")
```
