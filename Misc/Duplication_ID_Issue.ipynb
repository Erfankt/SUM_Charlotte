```python
import geopandas as gpd,numpy as np,pandas as pd, folium
```

```python
df = gpd.read_file("../../../../../../Edited Subdivisions Zurikanen Erfan 4-20=2025/subs_plus_new/sub.gdb",layer='final_10_9_25')
```

```python
# Step 1: Identify duplicated rows
dupes = df[df.duplicated(subset=['subd_id'], keep=False)].copy();dupes
```

```python
# Step 2: Split duplicates in half
def mark_half(group):
    half = len(group) // 2   # floor division
    group.loc[group.index[:half], 'replace'] = True
    group.loc[group.index[half:], 'replace'] = False
    return group

dupes = dupes.groupby('subd_id', group_keys=False).apply(mark_half)

# Step 3: Create a pool of new unique IDs
max_id = df['subd_id'].max()
n_new = int(dupes['replace'].sum())  # make sure it's an integer
new_ids = list(range(int(max_id) + 1, int(max_id) + 1 + n_new))

# Step 4: Assign new unique IDs only to the marked half
dupes.loc[dupes['replace'] == True, 'subd_id'] = new_ids

# Step 5: Update original DataFrame
df.update(dupes[['subd_id']])
```

```python
df['subd_id'] = df['subd_id'].astype(int)
```
